```python
f_mcpb = open("/home/leiqian/Documents/GROMACS_FOXO1/FEP/FOXO1_FEP/3COA_IRE/topol_combine/3COA_dry_GMX.top", "r")
```

```python
atomtypes_mcpb_list = []
atomtypes_mcpb_liststr = []

while True:
    
    f_mcpb_line = f_mcpb.readline()
    ## strip() makes space line become empty line   
    f_mcpb_listPerLine = f_mcpb_line.strip().split()    
    if len(f_mcpb_listPerLine) <= 2:
        continue
        
    if f_mcpb_listPerLine[1] == "atomtypes":
        atomtypes_mcpb_list.append(f_mcpb_listPerLine)
        atomtypes_mcpb_liststr.append(f_mcpb_line)
        break

while True:
    
    f_mcpb_line = f_mcpb.readline()
    f_mcpb_listPerLine = f_mcpb_line.strip().split()
    if len(f_mcpb_listPerLine) <= 2: 
        break        
    atomtypes_mcpb_list.append(f_mcpb_listPerLine)
    atomtypes_mcpb_liststr.append(f_mcpb_line)

        
f_mcpb.close()
```

```python
print(len(atomtypes_mcpb_list))
print(atomtypes_mcpb_list)
print(atomtypes_mcpb_liststr)
```

```python
f_atomtypes = open("/home/leiqian/Documents/GROMACS_FOXO1/FEP/FOXO1_FEP/3COA_IRE/topol_combine/new1_atomtypes.itp", "w")

for i in range(0, 37, 1):
    
    new_str = atomtypes_mcpb_liststr[i]

    f_atomtypes.write(new_str)

f_atomtypes.write("\n")
    
f_atomtypes.close()

# Done!

```
